# [dbo].[CRE_DB_BACKUP]

```sql
/****** Object:  StoredProcedure [dbo].[CRE_DB_BACKUP]    Script Date: 2025/11/19 18:30:50 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[CRE_DB_BACKUP]

    @BackupPath NVARCHAR(500)
AS
BEGIN
    SET NOCOUNT ON;

    DECLARE @BackupQuery NVARCHAR(MAX);

    -- バックアップクエリを生成
    SET @BackupQuery = N'BACKUP DATABASE [SUPPORTDB] TO DISK = ''' + @BackupPath + ''' WITH NOFORMAT, NOINIT, NAME = ''Full Backup'', SKIP, NOREWIND, NOUNLOAD, STATS = 10;';

    -- 動的SQLを実行
    EXEC sp_executesql @BackupQuery;
    
    PRINT 'バックアップが完了しました。';
END;
GO
```

# [dbo].[DEL_D_作業報告]

```sql
/****** Object:  StoredProcedure [dbo].[DEL_D_作業報告]    Script Date: 2025/11/19 18:30:50 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[DEL_D_作業報告]

	@SEQNO AS bigint,
	@削除日時 AS datetime,	
	@更新マシン AS nvarchar(255),	
		
	@OUTPUTSEQNO int output

AS
SET NOCOUNT ON;

SET @OUTPUTSEQNO = 0

IF  EXISTS(SELECT * FROM D_作業報告
			WHERE SEQNO = @SEQNO)
			
	--修正登録
	BEGIN
	UPDATE D_作業報告
	SET
	    削除日時 = @削除日時
	  , 更新マシン = @更新マシン
	WHERE SEQNO = @SEQNO
	
	SET @OUTPUTSEQNO = @SEQNO

	END

GO
```

# [dbo].[UPD_SYS_管理]

```sql
/****** Object:  StoredProcedure [dbo].[UPD_SYS_管理]    Script Date: 2025/11/19 18:30:50 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[UPD_SYS_管理]
	@入力マシン varchar(50)
    , @フィールド名 varchar(50)
	, @値 varchar(50)
AS

SET NOCOUNT ON;

DECLARE @sql NVARCHAR(MAX);
DECLARE @params NVARCHAR(MAX);

IF EXISTS(SELECT * FROM SYS_管理
			WHERE 入力マシン = @入力マシン)
	BEGIN
		-- 動的SQLをパラメータ化して構築
		SET @sql = N'UPDATE SYS_管理 SET [' + @フィールド名 + N'] = @値 WHERE 入力マシン = @入力マシン';
		SET @params = N'@値 varchar(50), @入力マシン varchar(50)';
		EXEC sp_executesql @sql, @params, @値 = @値, @入力マシン = @入力マシン;
	END

ELSE

	BEGIN
		-- 挿入の場合も動的SQLをパラメータ化
		SET @sql = N'INSERT INTO SYS_管理 (入力マシン, [' + @フィールド名 + N']) VALUES (@入力マシン, @値)';
		SET @params = N'@値 varchar(50), @入力マシン varchar(50)';
		EXEC sp_executesql @sql, @params, @値 = @値, @入力マシン = @入力マシン;
	END
GO
```
